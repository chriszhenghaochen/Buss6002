## New Visualizations
### Histogram

```python
%matplotlib inline
import matplotlib.pyplot as plt

# find out the 'AmountSpent' column
amount_spent = marketing['AmountSpent']

# plot the histogram of amount 
plt.ylabel('Frequency/ How many people')
plt.xlabel('Amount_spent')

plt.hist(amount_spent, bins = 100)

plt.show()
```

### Boxplot

```python
# gives you the colums of AmountSpent and Salary
columns = marketing[['AmountSpent', 'Salary']]

# columns.std() computes the standard deviation
# / columns.std() means try to re-normalize the limit to be 1 
std_columns = marketing[['AmountSpent', 'Salary']] / columns.std()

# std_columns.T: this to to get transpose
std_columns.transpose()

# Create the box plot
plt.boxplot(std_columns.transpose())

# Label the xticks
plt.xticks([1, 2], ["Amount Spent", "Salary"])

plt.show()
```

```python
# Q3:
import matplotlib.pyplot as plt

# think about what this do and modify the original data frame

# 1. family = (marketing['Married'] == 'Married') & (marketing['Children'] >= 1) is to create a true/false column
# 2. marketing['Family'] = family: is to insert the family into data frame
marketing['Family'] = (marketing['Married'] == 'Married') & (marketing['Children'] >= 1)

# think about the this weird condition and selection and plot
# don't worry about the alpha, it is just tranparency :)

# 1. family_rows = marketing[marketing['Family'] == True] is to return all rows meets the 'family' condition or the cell here is True
# 2. family_rows['Salary'], select the 'Salary' column from the family_rows
# 3. plt.hist(family_rows['Salary']) is to plot the histogram of family_rows['Salary']

plt.hist(marketing[marketing['Family'] == True]['Salary'], alpha = 0.5, label = 'Family Salary', bins = 20)
plt.hist(marketing[marketing['Family'] == False]['Salary'], alpha = 0.5, label = 'Non-Family Salary', bins = 20)

plt.ylabel('Frequency')
plt.xlabel('Salary')
plt.legend()
plt.title('Salary for family or non-family')
plt.show()
```
